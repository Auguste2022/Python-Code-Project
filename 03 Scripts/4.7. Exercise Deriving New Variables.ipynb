2. Create condition to include filters

```python
# Defined a function
def price_label(row):
    
    if row['prices']<= 5:
        return 'Low-range product'
    elif(row['prices'] < 5) and (row['prices']<= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
# Apply the function
df_orders_products_merged['price_range'] = df_orders_products_merged.apply(price_label, axis=1)
```

```python
df_orders_products_merged['price_range'].value_counts(dropna = False)
```

4.Creating condition if-else use loc.

```python
df_orders_products_merged.loc[df_orders_products_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_orders_products_merged.loc[(df_orders_products_merged['prices'] <= 15) & (df_orders_products_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df_orders_products_merged.loc[df_orders_products_merged['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

5. Print the frequency of the new column

7. Print the frequency of the new column

```python
result = []

for value in df_orders_products_merged["orders_day_of_wek"]:
  if value == 0 or value == 1:
    result.append("Busiest days")
  elif value == 3 or value == 4:
    result.append("slowest days")
  else:
    result.append("Regular days")
```

```python
result
```

8.Create a new column for the busiest and slowest days

```python
df_orders_products_merged['Busiest_days'] = result
```

```python
df_orders_products_merged['Busiest_days'].value_counts(dropna = False)
```


Observations:

"Regular days" have the highest count: This category has the most occurrences, with 400,720. This suggests that a significant number of days fall into this category.

"Busiest days" have the second-highest count: "Busiest days" are close in count to "Regular days," with 358,601 occurrences. This indicates that there are periods when order activity is high, but they are not as frequent as regular days.

"Slowest days" have the lowest count: "Slowest days" have the fewest occurrences, with 240,679. This implies that there are specific days or periods with relatively low order activity.
