2. Creatte a loyality flag loc

```python
# Aggregating data with transform().
# Transform() function create a new column containing maximum frequency of the order_number column.
# then using loc function, to create second column containing a flag designating whether a customer is loyal or not.
# 1.split data into groups based on the "user_id" column
# 2.apply the transformation()function on the "order_number" column to generate the maximun orders for each user.
# 3.create a new column, "max_order", into which you'will place the results of the aggregation.
df_orders_products_merged['max_order'] = df_orders_products_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

The results for the entire dataframe and the subset dataframe both provide the mean values of the 'order_number' column grouped by 'department_id'. The main difference is the scope of the data being analyzed.

Subset Dataframe: The subset dataframe appears to provide the mean 'order_number' values for each 'department_id' within a specific subset of the data.
Entire Dataframe: The entire dataframe seems to provide the mean 'order_number' values for each 'department_id' across the entire dataset.

In both cases, the mean values for 'order_number' are calculated for different departments, but the subset dataframe represent a specific subset of the data (possibly a subset of departments) compared to the entire dataframe, which considers all departments.

3. Create a spending flag for each user based on the average price across all their orders

```python
average_order_price = df_orders_products_merged.groupby('user_id')['prices'].mean()
```

```python
df_orders_products_merged['spending_flag'] = df_orders_products_merged['user_id'].map(average_order_price.apply(lambda mean: 'Low spender' if mean < 10 else 'High spender'))
```

```python
df_orders_products_merged.head()
```

4. Create an order frequency flag for customers based on the median of "days_since_prior_order"

```python
median_days_since_prior_order = df_orders_products_merged.groupby('user_id')['days_since_prior_order'].median()
```

```python
df_orders_products_merged['frequency_flag'] = df_orders_products_merged['user_id'].map(median_days_since_prior_order.apply(lambda median: "Non-frequent customer" if median > 20 else ("Regular customer" if median > 10 else "Frequent customer")))
```
